# Get Friend-Enemy Edge List

```python
import itertools


def construct_ally_enemy_pairs(c_e_id, c_id = None, e_id = None):
    
    ally_enemy_pairs = list()

    if c_id != None:
        c_e_id = {k: v for (k,v) in c_e_id.items() if k in c_id}


    for c_id in list(c_e_id.keys()):
        combatants = c_e_id[c_id]
        #n_bell = len(combatants)
        #all_entities = list(itertools.chain(*combatants))

        ## enemies
        for combatant_pair in itertools.combinations(combatants, r=2):
            for enemy_pair in itertools.product(*combatant_pair):
                if enemy_pair[0] < enemy_pair[1]:
                    enemy_pair_sorted = (enemy_pair[0], enemy_pair[1])
                else:
                    enemy_pair_sorted = (enemy_pair[1], enemy_pair[0])
                #ally_enemy_pairs.append((enemy_pair[0], enemy_pair[1], c_id, "enemies"))
                ally_enemy_pairs.append(enemy_pair_sorted)

        for alliance in combatants:

            ## allies
            for friend_pair in itertools.combinations(alliance, r=2):
                if friend_pair[0] < friend_pair[1]:
                    friend_pair_sorted = (friend_pair[0], friend_pair[1])
                else:
                    friend_pair_sorted = (friend_pair[1], friend_pair[0])
                    
                #ally_enemy_pairs.append((friend_pair[0], friend_pair[1], c_id, "allies"))
                ally_enemy_pairs.append(friend_pair_sorted)

    return ally_enemy_pairs


c_e_id = dict(zip(map_df["conflict_id"], map_df["entity_ids"]))
ally_enemy_list = construct_ally_enemy_pairs(c_e_id)
#ally_enemy_df = pd.DataFrame(ally_enemy_list, columns = ['entity_id1', 'entity_id2', 'conflict_id', 'type']) 
```
